```python
import sys
import delivery_driver as dd
sys.path.append("../../deliveryrobot")
from utilities.utilities import *
from kinematics.movementai import *
import time
import math
```

```python
start_time = time.time()
robot = dd.DeliveryRobot()
print(f"Robot setup complete in {time.time()-start_time} seconds")

target = math.pi/4
```

```python
robot.target_ai.slam_update([0.,0.,1.7])

goal_time = 1
start_time = time.time()
last_call = time.time()
call_time = 0
total_time = 0
while robot.align_ai(goal_time, call_time):
    
    # get time it took to make the call
    call_time = time.time() - last_call
    last_call = time.time()
    total_time+=call_time
    
    if total_time > 4:
        robot.robot.stop()
        
        break
        
robot.robot.stop()
robot.robot_ai.kplot.plot_data()
target += math.pi/4
```
